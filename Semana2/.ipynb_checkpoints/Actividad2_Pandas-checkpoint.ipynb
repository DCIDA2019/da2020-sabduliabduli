```python
#Unir datos de temperaturas máximas, mínimas y medias del 2019.
#Importamos datos de cada uno los archivos:

Temp_max_2019 = pd.read_csv(filepath_or_buffer = "201901010000TMax.csv", engine = 'python')
Temp_med_2019 = pd.read_csv(filepath_or_buffer = "201901010000TMed.csv", engine = 'python')
Temp_min_2019 = pd.read_csv(filepath_or_buffer = "201901010000TMin.csv", engine = 'python')

#Verificamos el formato (nombres) de cada uno de las columnas de los DataFrames:
print(Temp_max_2019.columns)
print(Temp_med_2019.columns)
print(Temp_min_2019.columns)
```

```python
#Ya que todas coinciden continuamos a concatenar todas en base a su "Clave".
#Primeramente creamos sub-DataFrames que sólo contegan las Claves y las temperaturas para evitar repetir datos si mezclamos
#todo junto.

#Aquí creamos un sub_DataFrame para la temperatura media únicamente con las columnas 'Clave' y 'Tmed'.
claves_Temp_med_2019 = pd.Series(Temp_med_2019['Clave'])
Tmed_Temp_med_2019 = pd.Series(Temp_med_2019['Tmed'])
sub_Temp_med_2019 = pd.DataFrame({'Clave': claves_Temp_med_2019, 'Tmed': Tmed_Temp_med_2019})

#Creamos un sub_DataFrame para la temperatura máxima únicamente con las columnas 'Clave' y 'Tmax'.
claves_Temp_max_2019 = pd.Series(Temp_max_2019['Clave'])
Tmax_Temp_max_2019 = pd.Series(Temp_max_2019['Tmax'])
sub_Temp_max_2019 = pd.DataFrame({'Clave': claves_Temp_max_2019, 'Tmax': Tmax_Temp_max_2019})

#Como 'merge()' sólo admite dos DataFrames y tenemos 3 primero hacemos "merge" con dos y de esa mezcla (#preliminar_data)
#resultante lo juntamos con el DataFrame que queda para así tener "final_data"
preliminar_data = pd.merge(Temp_min_2019, sub_Temp_med_2019, on = 'Clave')
final_data = pd.merge(preliminar, sub_Temp_max_2019, on = 'Clave')
final_data

#Finalmente obtenemos en un sólo DataFrame las 3 temperaturas min, med y max del 2019.
```
