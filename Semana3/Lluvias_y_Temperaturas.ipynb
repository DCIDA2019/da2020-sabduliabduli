```python
#Saúl Alí Caudillo Rodríguez
#DA2020

%pylab
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

#Extracción de datos desde .csv
lluvias_2018 = pd.read_csv(filepath_or_buffer = '201801010000Lluv.csv',
                           engine = 'python')
lluvias_2019 = pd.read_csv(filepath_or_buffer = '201901010000Lluv.csv',
                          engine = 'python')
max_temp_2018 = pd.read_csv(filepath_or_buffer = '201801010000TMax.csv',
                           engine = 'python')
max_temp_2019 = pd.read_csv(filepath_or_buffer = '201901010000TMax.csv',
                            engine = 'python')


#Datos de lluvias para el año 2018.
lon_Lluv_18 = lluvias_2018['LON'].values
lat_Lluv_18 = lluvias_2018['LAT'].values
ene_18 = lluvias_2018['ENE'].values

#Datos de lluvias para el año 2019.
lon_Lluv_19 = lluvias_2019['LON'].values
lat_Lluv_19 = lluvias_2019['LAT'].values
ene_19 = lluvias_2019['ENE'].values

#Datos de Temperaturas máximas 2018
lon_Temp_18 = max_temp_2018['Lon'].values
lat_Temp_18 = max_temp_2018['Lat'].values
max_temps_18 = max_temp_2018['Tmax'].values

#Datos de Temperaturas máximas 2019
lon_Temp_19 = max_temp_2019['Lon'].values
lat_Temp_19 = max_temp_2019['Lat'].values
max_temps_19 = max_temp_2019['Tmax'].values

#Creación de mapa de fondo centrado en México.
fig = plt.figure(figsize(10,10))
m = Basemap(projection = 'lcc', resolution = None,
           lat_0 = 25, lon_0 = -101,
           width = 4E6, height = 2.5E6)
m.shadedrelief(alpha = 0.9)

#Scattering
m.scatter(lon_Lluv_18, lat_Lluv_18, latlon = True, c = np.log10(ene_18),
          cmap = 'Blues', alpha = 0.7, s = 30)
plt.colorbar(label=r'$log_{10}(Precipitación)$ (mm)')
plt.clim(-1,3)
plt.title('Precipitación acumulada mensual: Enero 2018')
plt.show()
```
